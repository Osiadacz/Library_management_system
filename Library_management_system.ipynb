```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox as mb
from tkinter import scrolledtext as st
import mysql.connector
```

```python
class Form:
    def __init__(self):
        self.book_1 = Items()
        self.window_1 = tk.Tk()
        self.window_1.title("Library manager")
        self.bookmark_1 = ttk.Notebook(self.window_1)
        self.book_request()
        self.book_list()
        self.bookmark_1.grid(column = 0, row = 0, padx = 10, pady = 10)
        self.window_1.mainloop()
        
    def book_request(self):
        self.page_1 = ttk.Frame(self.bookmark_1)
        self.bookmark_1.add(self.page_1, text = "Book request by ID")
        self.label_frame_1 = ttk.LabelFrame(self.page_1, text = "Book")
        self.label_frame_1.grid(column = 0, row = 0, padx = 20, pady = 20)
        self.label_1 = ttk.Label(self.label_frame_1, text = "Book ID:")
        self.label_1.grid(column = 0, row = 0, padx = 4, pady = 4)
        self.id = tk.StringVar()
        self.entry_id = ttk.Entry(self.label_frame_1, textvariable = self.id)
        self.entry_id.grid(column = 1, row = 0, padx = 4, pady = 4, sticky=tk.W)
        self.label_2 = ttk.Label(self.label_frame_1, text = "Book name:")
        self.label_2.grid(column = 0, row = 1, padx = 4, pady = 4)
        self.book_name = tk.StringVar()
        self.entry_book_name = ttk.Entry(self.label_frame_1, textvariable = self.book_name, state = "readonly")
        self.entry_book_name.grid(column = 1, row= 1, padx = 4, pady = 4, sticky=tk.W, ipadx = 100)
        self.label_3 = ttk.Label(self.label_frame_1, text = "Author:")
        self.label_3.grid(column = 0, row = 2, padx = 4, pady = 4)
        self.author = tk.StringVar()
        self.entry_author = ttk.Entry(self.label_frame_1, textvariable = self.author, state = "readonly")
        self.entry_author.grid(column = 1, row = 2, padx = 4, pady = 4, sticky=tk.W)
        self.label_4 = ttk.Label(self.label_frame_1, text = "Description:")
        self.label_4.grid(column = 0, row = 3, padx = 4, pady = 4)
        self.description = tk.StringVar()
        self.entry_description = ttk.Entry(self.label_frame_1, textvariable = self.description, state = "readonly")
        self.entry_description.grid(column = 1, row = 3, padx = 4, pady = 4, ipadx = 400)
        self.label_5 = ttk.Label(self.label_frame_1, text = "Year published:")
        self.label_5.grid(column = 0, row = 4, padx = 4, pady = 4)
        self.year = tk.StringVar()
        self.entry_year = ttk.Entry(self.label_frame_1, textvariable = self.year, state = "readonly")
        self.entry_year.grid(column = 1, row = 4, padx = 4, pady = 4, sticky=tk.W)
        self.label_6 = ttk.Label(self.label_frame_1, text = "Availability:")
        self.label_6.grid(column = 0, row = 5, padx = 4, pady = 4)
        self.availability = tk.StringVar()
        self.entry_availability = ttk.Entry(self.label_frame_1, textvariable = self.availability, state = "readonly")
        self.entry_availability.grid(column = 1, row = 5, padx = 4, pady = 4, sticky=tk.W)
        self.label_7 = ttk.Label(self.label_frame_1, text = "Loan system:")
        self.label_7.grid(column = 0, row = 6, padx = 4, pady = 4)
        self.loan_system = tk.StringVar()
        self.entry_loan_system = ttk.Entry(self.label_frame_1, textvariable = self.loan_system, state = "readonly")
        self.entry_loan_system.grid(column = 1, row = 6, padx = 4, pady = 4, sticky=tk.W)
        self.label_8 = ttk.Label(self.label_frame_1, text = "Days of the loan:")
        self.label_8.grid(column = 0, row = 7, padx = 4, pady = 4)
        self.days_loan = tk.StringVar()
        self.entry_days_loan = ttk.Entry(self.label_frame_1, textvariable = self.days_loan, state = "readonly")
        self.entry_days_loan.grid(column = 1, row = 7, padx = 4, pady = 4, sticky=tk.W)
        self.button_1 = ttk.Button(self.label_frame_1, text = "Request", command = self.request)
        self.button_1.grid(column = 1, row = 8, padx = 4, pady = 4)

        
    def request(self):
        data = (self.id.get(), )
        answer = self.book_1.requested(data)
        if len(answer) > 0:
            self.book_name.set(answer[0][0])
            self.author.set(answer[0][1])
            self.description.set(answer[0][2])
            self.year.set(answer[0][3])
            self.availability.set(answer[0][4])
            self.loan_system.set(answer[0][5])
            self.days_loan.set(answer[0][6])
        else:
            self.book_name.set('')
            self.author.set('')
            self.description.set('')
            self.year.set('')
            self.availability.set('')
            self.loan_system.set('')
            self.days_loan.set('')
            mb.showinfo("Information", "There is no book with that code")
        
        
    
    def book_list(self):
        self.page_2 = ttk.Frame(self.bookmark_1)
        self.bookmark_1.add(self.page_2, text = "Full list of books")
        self.label_frame_2 = ttk.LabelFrame(self.page_2, text = "Book")
        self.label_frame_2.grid(column = 0, row = 0, padx = 30, pady = 35)
        self.button_1 = ttk.Button(self.label_frame_2, text = "Full list of books", command = self.show)
        self.button_1.grid(column = 0, row = 0, padx = 4, pady = 4)
        self.scrolled_text = st.ScrolledText(self.label_frame_2, width = 40, height = 8)
        self.scrolled_text.grid(column = 0, row = 1, padx = 30, pady = 35)
        
    
    def show(self):
        answer = self.book_1.get_all()
        self.scrolled_text.delete("1.0", tk.END)
        for row in answer:
            self.scrolled_text.insert(tk.END, "ID:"+str(row[0])+
                                              "\nBook:"+row[1]+
                                              "\nAuthor:"+str(row[2])+"\n\n")
        
    

class Items:
    
    def open(self):
        connection = mysql.connector.connect(host = "localhost",
                                                user = "root",
                                                password = "2204William1401",
                                                database = "library")
        
        return connection
    
    
    def requested(self, data):
        conn = self.open()
        cursor = conn.cursor()
        sql = "select book_name, author, description, year_published, availability, loan_system, days_loan from library_management where id = %s"
        cursor.execute(sql, data)
        conn.close()
        return cursor.fetchall()
    
    
    def get_all(self):
        conn = self.open()
        cursor = conn.cursor()
        sql = "select id, book_name, author from library_management"
        cursor.execute(sql)
        conn.close()
        return cursor.fetchall()
    
    

app = Form()
```
